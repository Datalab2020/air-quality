```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
url = "https://opendata.arcgis.com/datasets/6f64bbd4f94c425791c2ec7eee33bb71_0.csv"
df = pd.read_csv(url)
df.isna().sum()
```

```python
df.date_ech = pd.to_datetime(df.date_ech, infer_datetime_format=True)
```

```python
df
```

```python
df_Or = df.iloc[np.where(df['lib_zone'] == 'BLOIS')[0], :]
```

```python
df_Or
```

```python
df['lib_zone'].unique()
```

```python
df_Or['mois'] = df_Or['date_ech'].dt.month
df_Or['annee'] = df_Or['date_ech'].dt.year

dfpm = df_Or.groupby(by=['annee', 'mois']).mean()

listPolluants = ['val_no2','val_so2','val_pm10','val_pm25']

dfpm.plot(y=listPolluants)
```

```python
df_Or = df.iloc[np.where(df['annee'] == '2020')[0], :]
df_Or
```

```python
for i in df['lib_zone'].unique() :
    df_Or = df.iloc[np.where(df['lib_zone'] == i)[0], :]
    df_Or['mois'] = df_Or['date_ech'].dt.month;
    df_Or['annee'] = df_Or['date_ech'].dt.year;
    dfpm = df_Or.groupby(by=['annee', 'mois']).mean()
    dfpm.plot(y=['val_no2','val_so2','val_pm10','val_o3','val_pm25'], figsize= (15,5))
    plt.title(i);
    
```
